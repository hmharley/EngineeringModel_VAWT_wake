## Gaussian VAWT wake model
This is a model proposed by Abkar(2018)

```python
x = np.linspace(1,6,50)
y = 0; z = 0; zh = 0
k_star = 0.04; Ct = 0.94
D = 1; H=1; k_starD=k_star; k_starH =k_star;
DU_U0,sigmaD,sigmaH = EM.GaussianVAWT(Ct, x,y,z,zh,D,H, k_starD, k_starH);
Ur_over_U0 = 1 - DU_U0


plt.plot(x,Ur_over_U0,'--r')
plt.show()
```

## Top-hat VAWT wake model
This is a model proposed by Abkar(2018)

```python
x = np.linspace(0,6,50)
Dr = 1;
Ct = 0.9;
H = 1;

k = 0.066;
kD = k;kH = k;
Ur_over_U0,Hw, Dw = EM.TophatVAWT(Ct,H,Dr,x,kD,kH)
plt.plot(x,Ur_over_U0,'--r')
plt.show()
```
